### Paso 1: Calcular la matriz TF-IDF para el corpus `data/01_corpus_turismo_500.txt`

```python
# ── Instalación de dependencias ──────────────────────────────────
!pip install scikit-learn pandas numpy --quiet
```

```python
# ── Imports ──────────────────────────────────────────────────────
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# ── Carga del corpus ─────────────────────────────────────────────
CORPUS_PATH = "data/01_corpus_turismo_500.txt"

with open(CORPUS_PATH, encoding="utf-8", errors="ignore") as f:
    documents = [line.strip() for line in f if line.strip()]

print(f"Documentos cargados: {len(documents)}")
print(f"Ejemplo: {documents[0][:100]}")
```

```python
# ── Cálculo de la matriz TF-IDF ──────────────────────────────────
tfidf_vectorizer = TfidfVectorizer(stop_words="english", max_features=1000)
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

feature_names = tfidf_vectorizer.get_feature_names_out()

df_tfidf = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=feature_names
)

print(f"Dimensiones de la matriz TF-IDF: {df_tfidf.shape}")
df_tfidf.head()
```

### Paso 2: Construir el corpus `Gutenberg 1000`

El corpus `Gutenberg 1000` es un corpus compuesto por 1000 libros de Gutenberg Project

```python
# ── Carga del corpus Gutenberg 1000 desde archivos locales ───────
import os

GUTENBERG_DIR = "data/gutenberg/data"

gutenberg_corpus = [
    {"id": filename, "text": open(os.path.join(GUTENBERG_DIR, filename), encoding="utf-8", errors="ignore").read()}
    for filename in sorted(os.listdir(GUTENBERG_DIR))
    if filename.endswith(".txt")
]

df_gutenberg = pd.DataFrame(gutenberg_corpus)

print(f"Documentos cargados: {len(df_gutenberg)}")
df_gutenberg.head()
```

### Paso 3: Calcular la matriz TF-IDF para el corpus `Gutenberg 1000`

```python
# ── Matriz TF-IDF para el corpus Gutenberg 1000 ──────────────────
tfidf_vectorizer_gutenberg = TfidfVectorizer(stop_words="english", max_features=1000)
tfidf_matrix_gutenberg = tfidf_vectorizer_gutenberg.fit_transform(df_gutenberg["text"])

feature_names_gutenberg = tfidf_vectorizer_gutenberg.get_feature_names_out()

df_tfidf_gutenberg = pd.DataFrame(
    tfidf_matrix_gutenberg.toarray(),
    index=df_gutenberg["id"],
    columns=feature_names_gutenberg
)

print(f"Dimensiones de la matriz TF-IDF Gutenberg: {df_tfidf_gutenberg.shape}")
df_tfidf_gutenberg.head()
```

### Paso 4: Programar una función `buscar()` para el corpus `Gutenberg 1000`

```python
# ── Función de búsqueda por similitud coseno (TF-IDF) ────────────
from sklearn.metrics.pairwise import cosine_similarity

def buscar(query: str, vectorizer, tfidf_mat, df_docs, top_k: int = 5) -> pd.DataFrame:
    """
    Busca los documentos más relevantes para una consulta.
    
    Parámetros:
        query     : consulta en texto libre
        vectorizer: TfidfVectorizer ya entrenado
        tfidf_mat : matriz TF-IDF del corpus
        df_docs   : DataFrame con columnas 'id' y 'text'
        top_k     : número de documentos a retornar
    
    Retorna:
        DataFrame con ranking de documentos por relevancia
    """
    query_vector = vectorizer.transform([query])
    scores       = cosine_similarity(query_vector, tfidf_mat).flatten()

    return (
        pd.DataFrame({"documento": df_docs["id"].values, "score": scores})
        .sort_values("score", ascending=False)
        .head(top_k)
        .reset_index(drop=True)
        .assign(rank=lambda df: df.index + 1)
        .set_index("rank")
    )

# ── Prueba de la función ─────────────────────────────────────────
query = "adventure on the sea"

resultados = buscar(query, tfidf_vectorizer_gutenberg, tfidf_matrix_gutenberg, df_gutenberg)

print(f"Top 5 resultados para: '{query}'")
resultados
```
